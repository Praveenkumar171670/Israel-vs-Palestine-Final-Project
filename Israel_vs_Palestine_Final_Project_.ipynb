```python
unique_values = Dataset.nunique()
print("Unique Values:\n", unique_values)
```

```python
Gender=Dataset.groupby('gender').size().reset_index(name='incident_count')
print(Gender)
```

```python
Dataset['event_location_region'].value_counts().reset_index()
```

```python
Dataset['event_location_region'].value_counts().plot(kind='pie', colors=sns.color_palette('Set1'),autopct='%1.1f%%', figsize=(10,6))
```

```python
citizenship_counts=Dataset['citizenship'].value_counts()
event_location_region=Dataset['event_location_region'].value_counts()
print(citizenship_counts)
print(event_location_region)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='gender', data=Dataset, hue='citizenship')
plt.title('Citizenships within the Region')
plt.show()
```

```python
hostilities_counts =Dataset[Dataset['took_part_in_the_hostilities']=='Yes']['citizenship'].value_counts()
no_hostilities_counts =Dataset[Dataset['took_part_in_the_hostilities']=='No']['citizenship'].value_counts()
print(hostilities_counts)
print(no_hostilities_counts)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='citizenship', data=Dataset)
```

```python
Dataset['type_of_injury'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='type_of_injury', data=Dataset)

plt.xlabel("Injured People")
plt.ylabel("Type of injury")
plt.title('Injury Types')
plt.show()
```
